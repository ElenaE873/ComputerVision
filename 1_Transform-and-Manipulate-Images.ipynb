#### Purposes for Image Processing

ie. what can be accomplished by way of image processing

- 1 - Visualization:  **determining objects that are not visible**
- 2 - Image sharpening and restoration: **creating better images**
- 3 - Image retrieval: **seek for the image of interest**
- 4 - Measurement of pattern: **measuring various objects**
- 5 - Image recognition: **distinguishing objects in an image**
