#### Matplotlib Hitograms

Parameters:
- Input array - ie. frequency
- and bins 

Process:
- successive elements in bin array act as the boundary of each bin

```python
#1 - Obtain the red color slice
red = image[:, :, 0]

#2 - Use the histogram function
plt.hist(red.ravel(), #to return a continuous flattened array of red values from slice
         bins=256)  #show the number of pixels for every single pixel value (0 to 256)
plt.title('Red Histogram')
plt.show()
```
